# Análise Exploratória dos Microdados do ENEM

**Objetivo:** Este notebook tem como finalidade explorar e visualizar os resultados processados pelo pipeline de dados do ENEM. O pipeline principal (`main.py`) é responsável por:

1.  Baixar os dados brutos do INEP.
2.  Realizar a limpeza e transformações necessárias usando PySpark.
3.  Executar análises agregadas (médias, correlações, etc.).
4.  Salvar os resultados processados em formato Parquet no HDFS.

Neste ambiente, vamos nos conectar ao mesmo ecossistema (Spark e HDFS) para carregar os resultados pré-calculados e criar visualizações para interpretar os achados.

## 1. Configuração da Sessão Spark

Primeiro, vamos criar uma sessão Spark a partir do Jupyter. A configuração deve apontar para o `Spark Master` e o `HDFS NameNode` que estão rodando nos nossos containers Docker.

```python
import findspark
findspark.init()

from pyspark.sql import SparkSession
from pyspark.sql import functions as F
import matplotlib.pyplot as plt
import seaborn as sns

spark = (
    SparkSession.builder
    .appName("Análise ENEM via Jupyter")
    .master("spark://spark-master:7077")
    .config("spark.hadoop.fs.defaultFS", "hdfs://namenode:8020")
    .config("spark.executor.memory", "2g")
    .config("spark.driver.memory", "2g")
    .config("spark.network.timeout", "600s")
    .config("spark.executor.heartbeatInterval", "60s")
    .config("spark.ui.enabled", "true")
    .getOrCreate()
)

print("Sessão Spark criada com sucesso!")
print("Versão do Spark:", spark.version)
print("Versão do PySpark:", spark.sparkContext.pythonVer)
```

## 2. Verificando a Conexão com o HDFS

Vamos listar os diretórios de resultados no HDFS para confirmar que o pipeline principal (`main.py`) foi executado e que temos acesso aos dados.

```python
HDFS_URI = "hdfs://namenode:8020"
RESULTADOS_PATH = f"{HDFS_URI}/user/enem/resultados"

fs = spark._jvm.org.apache.hadoop.fs.FileSystem.get(spark._jsc.hadoopConfiguration())
path = spark._jvm.org.apache.hadoop.fs.Path(RESULTADOS_PATH)

print(f"Listando conteúdo de {RESULTADOS_PATH}:")
for f in fs.listStatus(path):
    print(f"- {f.getPath().getName()}")
```

## 3. Análise dos Resultados

Agora, vamos carregar cada um dos datasets de resultados (salvos em Parquet) e criar visualizações para entender os padrões.

### 3.1. Média de Matemática por UF

```python
df_uf = spark.read.parquet(f"{RESULTADOS_PATH}/media_por_uf").toPandas()

plt.figure(figsize=(14, 8))
sns.barplot(data=df_uf.sort_values("MEDIA_MT", ascending=False), x="MEDIA_MT", y="SG_UF_PROVA", palette="viridis")
plt.title("Média em Matemática por Estado", fontsize=16)
plt.xlabel("Média da Nota em Matemática", fontsize=12)
plt.ylabel("Estado", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

### 3.2. Média de Matemática por Tipo de Escola

O campo `TP_ESCOLA` significa:
- 1: Não Respondeu
- 2: Pública
- 3: Privada

```python
df_escola = spark.read.parquet(f"{RESULTADOS_PATH}/media_por_escola").toPandas()

# Mapeando os códigos para rótulos mais claros
escola_map = {1: 'Não Respondeu', 2: 'Pública', 3: 'Privada'}
df_escola['TIPO_ESCOLA_LABEL'] = df_escola['TP_ESCOLA'].map(escola_map)

plt.figure(figsize=(8, 5))
sns.barplot(data=df_escola.sort_values("MEDIA_MT"), x="TIPO_ESCOLA_LABEL", y="MEDIA_MT", palette="magma")
plt.title("Média em Matemática por Tipo de Escola", fontsize=16)
plt.xlabel("Tipo de Escola", fontsize=12)
plt.ylabel("Média da Nota", fontsize=12)
plt.show()
```

### 3.3. Correlação: Renda Familiar vs. Nota em Matemática

```python
df_corr = spark.read.parquet(f"{RESULTADOS_PATH}/correlacao_renda").toPandas()
correlation = df_corr.iloc[0]['CORR_Q006_MT']

print(f"Correlação de Pearson (Renda x Nota MT) em 2020: {correlation:.4f}")

plt.figure(figsize=(6, 4))
sns.barplot(x=["2020"], y=[correlation], palette=["rocket"])
plt.title("Correlação entre Renda e Nota em Matemática", fontsize=14)
plt.ylabel("Coeficiente de Pearson", fontsize=12)
plt.ylim(0, 1)
plt.text(0, correlation / 2, f'{correlation:.4f}', ha='center', color='white', fontsize=14, fontweight='bold')
plt.show()

print("Uma correlação forte e positiva indica que, em geral, maiores rendas familiares estão associadas a maiores notas.")
```

### 3.4. Análise de Desigualdade Regional

```python
df_regiao = spark.read.parquet(f"{RESULTADOS_PATH}/desigualdade_regiao").toPandas()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico de Média por Região
sns.barplot(data=df_regiao.sort_values("MEDIA_MT", ascending=False), x="MEDIA_MT", y="REGIAO", ax=ax1, palette="crest")
ax1.set_title("Média da Nota de MT por Região", fontsize=14)
ax1.set_xlabel("Média da Nota", fontsize=12)
ax1.set_ylabel("Região", fontsize=12)

# Gráfico de Desvio Padrão por Região
sns.barplot(data=df_regiao.sort_values("STD_MT", ascending=False), x="STD_MT", y="REGIAO", ax=ax2, palette="flare")
ax2.set_title("Dispersão (Desvio Padrão) da Nota de MT por Região", fontsize=14)
ax2.set_xlabel("Desvio Padrão da Nota", fontsize=12)
ax2.set_ylabel("", fontsize=12)

plt.tight_layout()
plt.show()

print("O desvio padrão indica a dispersão das notas. Uma região com maior desvio padrão possui uma maior desigualdade de desempenho entre seus estudantes.")
```

### 3.5. Média por Faixa de Renda

Visualizando o desempenho médio dos estudantes agrupados por diferentes faixas de renda familiar.

```python
df_faixa = spark.read.parquet(f"{RESULTADOS_PATH}/media_por_faixa_renda").toPandas()

# Ordenando as faixas de renda para o gráfico
faixa_order = ['Até 1k', '1k–3k', '3k–6k', 'Acima de 6k']
df_faixa['FAIXA_RENDA'] = pd.Categorical(df_faixa['FAIXA_RENDA'], categories=faixa_order, ordered=True)
df_faixa = df_faixa.sort_values('FAIXA_RENDA')

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_faixa, x='FAIXA_RENDA', y='MEDIA_MT', marker='o', markersize=8, color='b')
plt.title('Desempenho em Matemática por Faixa de Renda', fontsize=16)
plt.xlabel('Faixa de Renda Mensal (R$)', fontsize=12)
plt.ylabel('Média da Nota em Matemática', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## 4. Conclusão

As análises visuais confirmam tendências frequentemente observadas nos dados do ENEM:

- **Impacto Socioeconômico:** A renda familiar e o tipo de escola (pública vs. privada) são fortes preditores do desempenho em matemática.
- **Desigualdade Regional:** Existe uma clara disparidade de desempenho entre as diferentes regiões do Brasil, com o Sudeste e o Sul apresentando as maiores médias.

As métricas de desempenho do pipeline (throughput, uso de CPU/RAM) podem ser consultadas nos logs do container `spark-job`, conforme exibido ao final da sua execução.
